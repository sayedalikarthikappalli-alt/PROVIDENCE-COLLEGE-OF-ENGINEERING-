```python
from sklearn.linear_model import Perceptron
marks = [[10,20,30],[12,24,32],[82,83,84],[92,93,94]]
result = ["fail","fail","pass","pass"]
model = Perceptron()
model.fit(marks,result)
res = model.predict([[67,82,83]])
print(res)
```

```python
from sklearn.linear_model import Perceptron
import pandas as pd
file = pd.read_csv("/content/irisexcel.csv")
dimensions = file[["sepal_length","sepal_width","petal_length","petal_width"]]
species = file["species"]
model = Perceptron()
model.fit(dimensions,species)
res = model.predict([[1.2,1.4,1.3,1.5]])
print(res)
```
